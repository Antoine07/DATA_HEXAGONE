```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataset = {f"tirage_{k}" : np.random.rand(100) for k in range(8) }
```

```python
# Approche objet
fig, ax = plt.subplots(4,2, figsize=(12, 8))
fig.tight_layout(pad=4.0)
colors = ['red', 'orange', 'pink', 'blue', 'black', 'brown', 'yellow', 'purple']

x = np.linspace(0,100, 100)
k = 0
for i in range(4):
    for j in range(2):
        ax[i, j].plot(x,dataset[f'tirage_{i+j}'], c=colors[k])
        ax[i,j].set_title(f'subplot {k + 1}')
        ax[i,j].set_ylabel('Y')
        ax[i,j].set_xlabel('X')
        k += 1

plt.show()
```

```python
# Approche objet
plt.subplots(4,2, figsize=(12, 8))
colors = ['red', 'orange', 'pink', 'blue', 'black', 'brown', 'yellow', 'purple']

x = np.linspace(0,100, 100)
k = 0
for i in range(4):
    for j in range(2):
        plt.subplot(4, 2, k + 1)
        plt.plot(x,dataset[f'tirage_{i+j}'], c=colors[k])
        plt.title(f'Figure {k+1}')
        plt.xlabel('axe x')
        plt.ylabel('axe y')        
        k += 1
        
plt.subplots_adjust(left=0.1,
                    bottom=2.0, 
                    right=1, 
                    top=3, 
                    wspace=0.4, 
                    hspace=1.0)
plt.show()
```
