```python

import pandas as pd
import matplotlib.pyplot as plt
bitcoin = pd.read_csv(
    '../Data/BTC-EUR.csv',
    index_col='Date',
    parse_dates=True
)

bitcoin.head()
```

```python
# 01 Exercice 
# print(bitcoin.index)
bitcoin.plot(figsize=(12,5))
```

```python
# 02 Exercice 
bitcoin['2021']['Close'].plot(figsize=(10,5))
```

```python
# 02 Exercice 
bitcoin['2000':'2021']['Close'].plot(figsize=(10,5))
```

```python
bitcoin['2021']['Close'].resample('M').plot(figsize=(10,5));
```

```python
bitcoin['2021']['Close'].resample('M').mean().plot(figsize=(10,5));
```

```python
bitcoin['2021']['Close'].resample('M').std().plot(figsize=(10,5));
```

```python
m = bitcoin['2021']['Close'].resample('W').aggregate(['mean', 'std', 'min', 'max'])
```

```python
plt.figure(figsize=(12, 8))

m['mean'].plot(label="Moyenne par semaine")
plt.fill_between(m.index, m['max'], m['min'], alpha=.2, label="Min max Week")

plt.legend()
plt.show()
```
