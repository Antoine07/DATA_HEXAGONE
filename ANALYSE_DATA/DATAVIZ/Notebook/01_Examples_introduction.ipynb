```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# cr√©e un vecteur de 100 points entre 0 et 20
x = np.linspace(0, 20, 100)
```

```python
# Attention il faut que x et y est les m√™mes dimensions pour tracer la courbe
plt.figure(figsize=(12,8))
plt.plot(x, x**2, c='red', lw=2, label="quadratique", ls="--")

# print( (x**2).shape )
# print( x.shape )

plt.title('Figure')

plt.xlabel('axe x')
plt.ylabel('axe y')

# permet uniquement d'afficher le label
plt.legend()
# plt.savefig('my_fig.png')
plt.show()
```

```python
plt.plot(x, x**3, c='red', lw=2, label="cubique")
```

```python
plt.figure(figsize=(12,8))

plt.show()
```

```python
plt.figure(figsize=(12,8))

# lw √©paisseur, ls style du trait
plt.plot(x, x**2, label="quadratique", c='red', lw=2, ls='--')
plt.plot(x, x**3, label="cubique", c='black', lw=1, ls=':')

plt.title('Figures')

plt.xlabel('axe x')
plt.ylabel('axe y')

# On doit appeler cette fonction pour tracer la courbe
plt.legend()
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, x**2, label="quadratique", c='red')

plt.subplot(2, 1, 2)
plt.plot(x, x**3, label="quadratique", c='black')
```

```python
# Approche objet
fig, ax = plt.subplots(2,1, sharex=True, figsize=(12, 8))

# ax est un np.array 
ax[0].set_title('subplot 1')
ax[0].plot(x, x**2, c='red')
ax[0].plot(x, x**3, c= 'orange')

ax[1].set_title('subplot 2')
ax[1].plot(x, x**3)

ax[0].set_ylabel('YLabel 0')
ax[1].set_ylabel('YLabel 1')

ax[0].set_xlabel('X 0')
ax[1].set_xlabel('X 1')

fig.tight_layout(pad=3.0)

plt.show()
```
