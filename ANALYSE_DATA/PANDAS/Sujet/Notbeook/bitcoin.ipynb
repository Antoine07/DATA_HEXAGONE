```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn

# intégration des graphiques dans le notebook
%matplotlib inline
```

```python
data_source_names = "../data/bitcoin.csv"

bitcoins = pd.read_csv( 
    f"{data_source_names}"
)
```

```python
bitcoins.head()
```

```python
# Casting de colonne pour effecuter
bitcoins['Prix'] = (bitcoins['Prix'].str.replace('1 BTC = ', '').str.replace('€', '')).astype(float)
```

```python
bitcoins['Date'] = pd.to_datetime(bitcoins['Date'])
```

```python
bitcoins.head()
```

```python
bitcoins.info()
```

```python
bitcoins[['Date', 'Prix']].plot(x="Date", y="Prix", figsize=(15,10))

plt.title('Bitcoin valuation')
plt.xlabel('Year')
plt.ylabel('number')
# Affiche la légende
plt.legend(title=" Bitcoinn");
```
